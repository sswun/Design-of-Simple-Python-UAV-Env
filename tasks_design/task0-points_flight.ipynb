我们先加载一个常规的无人机，并赋予一些初始参数。  
在这些参数中：
- name表示无人机的名称；  
- type表示无人机类型；
- max_speed表示无人机的最大速度；
- max_range表示无人机的最大搜索范围；
- max_altitude表示无人机的最大飞行高度；
- max_power表示无人机的最大电池容量（电源容量）；
- max_load表示无人机的最大负载；
- k_up表示无人机向上飞行的消耗系数；
- k_down表示无人机向下飞行的消耗系数；
- k_around表示无人机环绕飞行的消耗系数；
- dt表示时间间隔。
- weights表示无人机的重量，指自重。

消耗系数与无人机的电池容量有关系，无人机单位时间内耗电量 = 消耗系数 * 无人机单位时间在对应方向飞行距离 * （自重+负载）
无人机类型type在执行特殊任务时会有作用。

```python
UAV1 = uav.Conventional_UAV(name='test1', type='0', max_speed=10,
                            max_range=100, max_altitude=10, max_power=100, max_load=10,
                            k_up=0.5, k_down=0.1, k_around=0.3, dt=0.01, weight=2)
```

无人机在任意时刻的属性都被保存在一个字典中，无人机执行飞行任务时的相关属性则被保存在另一个属性中。
